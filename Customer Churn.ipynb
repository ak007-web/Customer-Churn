**Relacing the black values into 0**

**To convert 0/1 values into 'Yes/No' values**
- 'Senior Citizen' column contains values into '0/1'
- we are converting the values from '0/1' to 'No/Yes'

```python
def con(value):
    if value==1:
        return "yes"
    else:
        return "no"
df['SeniorCitizen']=df['SeniorCitizen'].apply(con)
```

# Visualizations

```python
plt.figure(figsize=(12,10))
ax=sns.countplot(x='Churn',data=df,hue="Churn")
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Count of Customers who Churn")
```

- This graph shows that 5174 customers didn't Chrun along with 1869 customers Chrun

**Chrun Distribution using Pie Chart**

```python
gb=df.groupby("Churn").agg({'Churn':"count"})
gb
plt.figure(figsize=(10,8))
gb.plot(kind='pie',labels=gb.index,autopct="%1.2f%%",subplots=True)
plt.title("Churn Distribution")
plt.ylabel("")
plt.show()
```

- According to the Chart, 73.46% peoples didn't Chrun

**Gender vs Chrun**

```python
plt.figure(figsize=(12,10))
ax=sns.countplot(x="gender",data=df,hue="Churn")
for bars in ax.containers:
    ax.bar_label(bars)
plt.title("Gender vs Chrun")
plt.show()
```

- this Graph shows the number of Males and Females who Chrun and not

**Senior Citizen vs Chrun**

```python
ax=sns.countplot(x="SeniorCitizen",data=df,hue="Churn")
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

- According to this graph, we can see how many Seniors Chrun and how didn't

```python
plt.figure(figsize=(12,10))
ax=sns.histplot(x="tenure",data=df,bins=72,hue="Churn")
for bars in ax.containers:
    ax.bar_label(bars)
```

- people who have used services for a long time have stayed and prople who have used our services 1 or 2 months have churned

```python
ax=sns.countplot(x="Contract",data=df,hue="Churn")
for bars in ax.containers:
    ax.bar_label(bars)
plt.show()
```

- People who have month-month contract are likely to churn then from those who have 1 or 2 years contract

```python
columns = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity', 
           'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']

# Number of subplots
num_cols = len(columns)

# Calculate rows and columns for the subplots grid
rows = (num_cols + 2) // 3  # 3 plots per row

# Create subplots
fig, axes = plt.subplots(rows, 3, figsize=(15, rows * 4))
axes = axes.flatten()  # Flatten the axes for easy iteration

# Loop through each column and create a countplot
for i, col in enumerate(columns):
    sns.countplot(data=df, x=col, ax=axes[i],hue='Churn')
    axes[i].set_title(f'Countplot of {col}')
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count')

# Remove any empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

-
The visualizations show count plots of customer subscriptions for various services, with churn (Yes/No) as a comparison. Services like PhoneService, InternetService, and TechSupport reveal notable differences between churned and retained customers. Features such as OnlineSecurity and StreamingTV indicate higher churn among customers who lack these services, suggesting that specific features may influence customer retention
